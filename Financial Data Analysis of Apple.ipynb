Step 2: Data Collection
Data Source: Collected historical stock data for Apple Inc. (AAPL) using the yfinance library.
Data Range: Downloaded stock data from January 1, 2015, to January 1, 2023.
Data Saved: The data was saved to a CSV file named apple_stock_data.csv.

```python
import yfinance as yf
import pandas as pd

try:
    print("Starting data collection...")
    # Downloading stock data for Apple (AAPL)
    apple_data = yf.download('AAPL', start='2015-01-01', end='2023-01-01')
    print("Data collection complete. Data downloaded:")
    print(apple_data.head())  # Print the first few rows of the data
    apple_data.to_csv('apple_stock_data.csv')
    print("Data saved to 'apple_stock_data.csv'.")
except Exception as e:
    print(f"An error occurred: {e}")
```

Missing Values Check: Checked for missing values in the dataset. No missing values were found in any of the columns.

```python
import pandas as pd

# Load the collected data
apple_data = pd.read_csv('apple_stock_data.csv', index_col='Date', parse_dates=True)

# Display the first few rows of the dataframe
print("Data loaded successfully:")
print(apple_data.head())

# Check for missing values
print("\nChecking for missing values:")
print(apple_data.isnull().sum())

# Drop rows with missing values (if any)
apple_data.dropna(inplace=True)

# Display the cleaned data
print("\nCleaned data:")
print(apple_data.head())

# Save the cleaned data to a new CSV file
apple_data.to_csv('cleaned_apple_stock_data.csv')
print("Cleaned data saved to 'cleaned_apple_stock_data.csv'.")
```

## Step 5.1: Closing Price Over Time

In this step, we will visualize the closing price of Apple Inc. (AAPL) over time. This graph will help us identify trends and patterns in the stock's performance.

- **Y-axis**: Closing Price
- **Objective**: To observe how the closing price has fluctuated over the specified date range.

The plot will provide insights into the stock's historical performance, highlighting any significant increases or decreases in price.


```python
import matplotlib.pyplot as plt

# Plotting the closing price over time
plt.figure(figsize=(12, 6))
plt.plot(df_stock.index, df_stock['Close'], label='Closing Price', color='blue')
plt.title('Apple Inc. (AAPL) Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid()
plt.show()
```

- **Rotation of X-ticks**: The dates on the x-axis are rotated for better readability.
- **Legend**: A legend is included to label the closing price.

## Step 6: 20-Day and 50-Day Moving Averages

In this step, we will calculate and visualize the 20-day and 50-day moving averages for the closing price of Apple Inc. (AAPL). Moving averages help to smooth out short-term price fluctuations and highlight longer-term trends in the stock's performance.

- **20-Day Moving Average**: Reflects the average closing price over the last 20 days.
- **50-Day Moving Average**: Reflects the average closing price over the last 50 days.
- **Objective**: To observe how these moving averages can indicate potential support and resistance levels as well as the overall trend direction.


```python
# Calculate 20-Day and 50-Day Moving Averages
df_stock['20-Day MA'] = df_stock['Close'].rolling(window=20).mean()
df_stock['50-Day MA'] = df_stock['Close'].rolling(window=50).mean()

# Plotting the closing price along with moving averages
plt.figure(figsize=(12, 6))
plt.plot(df_stock.index, df_stock['Close'], label='Closing Price', color='blue', alpha=0.5)
plt.plot(df_stock.index, df_stock['20-Day MA'], label='20-Day Moving Average', color='red')
plt.plot(df_stock.index, df_stock['50-Day MA'], label='50-Day Moving Average', color='green')
plt.title('Apple Stock Closing Price and Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
```

### Graph Attributes

- **Figure Size**: The figure size is set to (12, 6) for a clear and readable visualization.
- **Line Plot**: The closing price, 20-day moving average, and 50-day moving average are plotted to compare short-term and long-term trends.
- **Title**: The title indicates the content of the graph.
- **X-axis (Date)**: Represents the timeline for the stock data.
- **Y-axis (Price)**: Shows the stock prices in USD.
- **Legend**: Identifies the closing price and moving averages with corresponding colors.
- **Grid**: A grid is added to aid in interpreting the price values on the graph.


## Step 7: Calculate and Plot Daily Returns

In this step, we will calculate the daily returns for Apple Inc. (AAPL) and visualize them to understand the stock's volatility over time. Daily returns are a useful measure for assessing how much the stock price has changed from one day to the next.

- **Daily Return Calculation**: Daily returns are calculated as the percentage change in the closing price from one day to the next.
- **Objective**: To visualize the daily returns, allowing us to identify periods of high volatility and assess the overall risk associated with the stock.


```python
# Calculate daily returns
df_stock['Daily Return'] = df_stock['Close'].pct_change()

# Plotting daily returns
plt.figure(figsize=(12, 6))
plt.plot(df_stock.index, df_stock['Daily Return'], color='purple', alpha=0.6)
plt.title('Apple Stock Daily Returns')
plt.xlabel('Date')
plt.ylabel('Daily Return (%)')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Line at y=0 for reference
plt.grid()
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
```

### Graph Attributes

- **Figure Size**: The figure size is set to (12, 6) for a clear and readable visualization.
- **Line Plot**: Daily returns are plotted to illustrate fluctuations in returns over time.
- **Title**: The title indicates the content of the graph.
- **X-axis (Date)**: Represents the timeline for the stock data.
- **Y-axis (Daily Return)**: Shows the daily return as a percentage.
- **Horizontal Reference Line**: A dashed line at y=0 indicates the level of no change in returns, helping to visualize gains and losses.
- **Grid**: A grid is added to aid in interpreting the daily return values.


## Step 8: Distribution of Daily Returns

In this step, we will visualize the distribution of daily returns for Apple Inc. (AAPL). This will help us understand how frequently certain levels of returns occur and assess the overall risk and volatility of the stock.

- **Objective**: To create a histogram of daily returns, allowing us to observe the distribution shape and identify any significant patterns in the returns.
- **Insight**: A normal distribution would indicate typical behavior, while skewness or heavy tails can indicate higher volatility or risk.


```python
# Plotting the distribution of daily returns
plt.figure(figsize=(12, 6))
plt.hist(df_stock['Daily Return'].dropna(), bins=50, color='lightblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Daily Returns for Apple Stock')
plt.xlabel('Daily Return (%)')
plt.ylabel('Frequency')
plt.axvline(0, color='red', linewidth=1, linestyle='--')  # Line at y=0 for reference
plt.grid()
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
```

### Histogram Attributes

- **Figure Size**: The figure size is set to (12, 6) for a clear and readable visualization.
- **Histogram**: Daily returns are represented in a histogram to visualize the frequency of different return values.
- **Title**: The title indicates what the graph represents.
- **X-axis (Daily Return)**: Shows the daily return as a percentage.
- **Y-axis (Frequency)**: Indicates how many times each return occurred in the dataset.
- **Vertical Reference Line**: A dashed red line at x=0 indicates the level of no change in returns, helping visualize the balance between gains and losses.
- **Bins**: The histogram uses 50 bins to show the distribution granularity.
- **Grid**: A grid is added to aid in interpreting the frequency values.
